```python
# 대상 점 찾기
import cv2

img = cv2.imread('road.jpg')

cv2.circle(img, (603, 400), 6, (0,255,0), cv2.FILLED)
cv2.circle(img, (665, 400), 6, (0,255,0), cv2.FILLED)
cv2.circle(img, (550, 800), 6, (0,255,0), cv2.FILLED)
cv2.circle(img, (1270, 760), 6, (0,255,0), cv2.FILLED)

cv2.imshow('road', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 원근법 적용 1
import cv2
import numpy as np

img = cv2.imread('road.jpg')

# output size 가로: 500, 세로: 800
WIDTH = 500
HEIGHT = 800

pts = np.array([[603, 400], [665, 400], [1270, 760], [550, 800]], dtype=np.float32)
dst = np.array([[0,0], [WIDTH, 0], [WIDTH, HEIGHT], [0, HEIGHT]], dtype=np.float32)

matrix = cv2.getPerspectiveTransform(pts, dst) # 좌표 변환행렬 반환 (공식이랄까)
result = cv2.warpPerspective(img, matrix, (WIDTH, HEIGHT))

cv2.imshow('Transformed Perspective', result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 회전된 이미지 올바로 세우기

```python
# 원근법 적용 2
import cv2
import numpy as np

img = cv2.imread('poker.jpg')

# output size 가로: 500, 세로: 800
WIDTH = 530
HEIGHT = 710

pts = np.array([[702, 143], [1133, 414], [726, 1007], [276, 700]], dtype=np.float32)
dst = np.array([[0,0], [WIDTH, 0], [WIDTH, HEIGHT], [0, HEIGHT]], dtype=np.float32)

matrix = cv2.getPerspectiveTransform(pts, dst) # 좌표 변환행렬 반환 (공식이랄까)
result = cv2.warpPerspective(img, matrix, (WIDTH, HEIGHT))

cv2.imshow('Transformed Perspective', result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
